```python
#CREATION COLONNE ID


# Créer un dictionnaire pour stocker les index correspondant à chaque valeur de "income"
index_dict = {}

# Parcourir le DataFrame et assigner un index unique pour chaque valeur de "income"
for index, row in data_cleaned.iterrows():
    income = row['income']
    if income not in index_dict:
        index_dict[income] = len(index_dict)  # Utilisation de len(index_dict) pour obtenir un index unique
    data_cleaned.at[index, 'ID'] = int(index_dict[income])  # Convertir en entier

# Convertir la colonne "ID" en type entier
data_cleaned['ID'] = data_cleaned['ID'].astype(int)
```
